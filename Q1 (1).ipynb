```python
import pandas as pd
import glob
import os
import re
import nltk
from nltk import word_tokenize
nltk.download('punkt')
nltk.download('stopwords')
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from nltk.stem import PorterStemmer
import numpy as np


```

```python
filenames = glob.glob('Humor,Hist,Media,Food/' + "/*.*")
```

```python
finalInvertedIndex = {}

index=0;
for file in filenames:
    with open(os.path.join("" + file), 'r', encoding='latin-1') as f:
        text = f.read()
#         print(text)
        text = text.lower()
        text = " ".join(text.split())
#         print("****************")
        text = re.sub(r'[^\w\s]', '', text)
        text = word_tokenize(text)
        tokenisedDocument = []
        en_stopwords = stopwords.words('english')
        for word in text:
            if word not in en_stopwords:
                tokenisedDocument.append(word)
                
        result = []
                
#         print("*******รท*22")
        for word in tokenisedDocument:
            result.append(PorterStemmer().stem(word))
        
#         print(result)
        for data in result:
#             print(data)
            if data in finalInvertedIndex:
                finalInvertedIndex[data].append(index)
            else:
                finalInvertedIndex[data] = [index]
    index=index+1

```
